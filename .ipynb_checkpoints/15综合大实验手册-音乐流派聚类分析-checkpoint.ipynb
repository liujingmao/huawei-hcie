3.数据预处理

```python
# 随机采样
data = origin_data.sample(10000).reset_index(drop=True)
data
```

```python
# 按类型存储列名
list_object = []
list_num = []
for col in data.columns:
    if data[col].dtypes=='object':
        # 对object类型的数据进行特征编码
        data[col] = pd.factorize(data[col])[0]
        list_object.append(col)
    else:
        list_num.append(col)
```

```python
data.head()
```

```python
# 特征缩放
from sklearn.preprocessing import StandardScaler

model_std = StandardScaler()
df_std = model_std.fit_transform(data[list_num])
df_std = pd.DataFrame(data=df_std,columns=list_num)
```

```python
df_std.head()
```

```python
# 合并数据
data[list_num] = df_std[list_num]
```

```python
data.head()
```

```python
# 独热编码
df_oh = pd.get_dummies(data)
df_oh.shape
```

```python
# 降维
from sklearn.decomposition import PCA
model_pca = PCA(n_components=20)
data_pca = model_pca.fit_transform(df_oh)
```

```python
data_pca
```

4.使用K-means算法建模

```python
from sklearn.cluster import KMeans
from sklearn import metrics
```

```python
k, silhouette,sse = [], [], []
for i in range(2,20):
    kmeans_sk = KMeans(n_clusters = i, init = 'k-means++', max_iter = 500, algorithm='elkan',n_init = 10,random_state = 0)
    kmeans_sk.fit(data)
    s1 = metrics.silhouette_score(data, kmeans_sk.labels_, metric='euclidean')  # 轮廓系数
    s2 = kmeans_sk.inertia_  #  sse(误差平方和)评分
    silhouette.append(s1)
    sse.append(s2)
    print(u'K = %s, Silhouette = %.03f, SSE = %.03f' % (i, s1, s2))
#画 silhouette
plt.figure()
plt.title('K-means silhouette')
plt.plot(range(2,20), silhouette, '*')
plt.plot(range(2,20) , silhouette , '-' , alpha = 0.5)
plt.xlabel('Number of Cluster')
plt.ylabel('Silhouette')
plt.show()
#画 sse
plt.figure()
plt.title('K-means SSE')
plt.xlabel('Number of Cluster')
plt.ylabel('SSE')
plt.plot(range(2,20), sse, 'o')
plt.plot(range(2,20) , sse , '-' , alpha = 0.5)
```

```python
k, silhouette,sse = [], [], []
for i in range(2,20):
    kmeans_sk = KMeans(n_clusters = i, init = 'k-means++', max_iter = 500, algorithm='elkan',n_init = 10,random_state = 0)
    kmeans_sk.fit(data_pca)
    s1 = metrics.silhouette_score(data_pca, kmeans_sk.labels_, metric='euclidean')  # 轮廓系数
    s2 = kmeans_sk.inertia_  #  sse(误差平方和)评分
    silhouette.append(s1)
    sse.append(s2)
    print(u'K = %s, Silhouette = %.03f, SSE = %.03f' % (i, s1, s2))
#画 silhouette
plt.figure()
plt.title('K-means silhouette')
plt.plot(range(2,20), silhouette, '*')
plt.plot(range(2,20) , silhouette , '-' , alpha = 0.5)
plt.xlabel('Number of Cluster')
plt.ylabel('Silhouette')
plt.show()
#画 sse
plt.figure()
plt.title('K-means SSE')
plt.xlabel('Number of Cluster')
plt.ylabel('SSE')
plt.plot(range(2,20), sse, 'o')
plt.plot(range(2,20) , sse , '-' , alpha = 0.5)
```

5.使用birch算法算法进行建模

```python
from sklearn.cluster import Birch
k, silhouette = [], []
for i in range(3, 12):
    Birch_model = Birch(n_clusters=i, threshold = 0.5).fit(data)
    k.append(i)
    s = metrics.silhouette_score(data, Birch_model.labels_, metric='euclidean')
    silhouette.append(s)
    print(u'K = %s, Silhouette = %.03f' % (i, s))
plt.figure()
plt.title('Birch')
plt.plot(k, silhouette, 'r')
plt.xlabel('number of cluster')
plt.ylabel('Silhouette')
```

6.使用SpectralClustering算法建模

```python
from sklearn.cluster import SpectralClustering
k, silhouette = [], []
for i in range(2, 12):
    spectral = SpectralClustering(n_clusters = i,affinity='nearest_neighbors',assign_labels='discretize').fit(data)
    k.append(i)
    result = spectral.labels_
    s = metrics.silhouette_score(data, result, metric='euclidean')
    silhouette.append(s)
    print(u'K = %s, Silhouette = %.03f' % (i, s))
plt.figure()
plt.plot(k, silhouette, 'r')
plt.title('SpectralClustering')
plt.xlabel('number of cluster')
plt.ylabel('silhouette')
```
