1.读取数据

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./Fish.csv')
df.head()
```

2.探索数据

```python
df.info()  # 需要对Species列进行label-encoding
```

```python
df['Species'] = pd.factorize(df['Species'])[0]  # 特征编码
```

5.拆分数据

```python
# 纵向拆分
X = df.drop(axis=1,columns='Species')
y = df['Species']
```

```python
from sklearn.preprocessing import StandardScaler
model_std = StandardScaler()
X_std = model_std.fit_transform(X)
X_std = pd.DataFrame(data=X_std,columns=X.columns)
```

```python
# 横向拆分
from sklearn.model_selection import  train_test_split

X_triain,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

6.建模

```python
from sklearn.linear_model import LinearRegression  # 线性回归

# 设置参数
model_lr = LinearRegression()
# 训练
model_lr.fit(X_triain,y_train)
# 评分
model_lr.score(X_test,y_test)
```

7.模型评估-偏差

```python
from sklearn.metrics import mean_absolute_error  # 平均绝对误差
from sklearn.metrics import mean_squared_error  # 均方误差
from sklearn.metrics import mean_squared_log_error   # 均方根误差
from sklearn.metrics import r2_score

y_pre = model_lr.predict(X_test)

mse = mean_squared_error(y_test,y_pre)
mae = mean_absolute_error(y_test,y_pre)
r2_s = r2_score(y_test,y_pre)
print(f'LinearRegression的mse为:{mse},mae为:{mae},r2为:{r2_s}')
```

```python
from sklearn.neighbors import KNeighborsRegressor   # 回归器

model_knn = KNeighborsRegressor()
model_knn.fit(X_triain,y_train)
```

```python
y_pre = model_knn.predict(X_test)

mse = mean_squared_error(y_test,y_pre)
mae = mean_absolute_error(y_test,y_pre)
r2_s = r2_score(y_test,y_pre)
print(f'KNeighborsRegressor的mse为:{mse},mae为:{mae},r2为:{r2_s}')
```

```python
# 随机森林
# GBDT
# xgBoost
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import AdaBoostRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsRegressor


model_rf = RandomForestRegressor()
model_gbdt = GradientBoostingRegressor()
model_xgb = XGBRegressor()

list_model = [model_rf,model_gbdt,model_xgb]
list_name = ['RandomForestRegressor','GradientBoostingRegressor','XGBRegressor']
for model,name in zip(list_model,list_name):   
    model.fit(X_triain,y_train)
    y_pre = model.predict(X_test)
    mse = mean_squared_error(y_test,y_pre)
    mae = mean_absolute_error(y_test,y_pre)
    r2_s = r2_score(y_test,y_pre)
    print(f'{name}的mse为:{mse},mae为:{mae},r2为:{r2_s}')
```

8.模型评估-方差

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer

# 交叉验证
cv_score = cross_val_score(model_knn,X,y,cv=5,scoring=make_scorer(mean_squared_error))
```

```python
cv_score
```

9.优化

```python
from sklearn.model_selection import GridSearchCV

parms= {'n_estimators':[10,20,30],'max_bin':[2,3,4]}
model_gs= GridSearchCV(estimator=model_xgb,param_grid=parms,scoring=make_scorer(mean_squared_error),n_jobs=3)
model_gs.fit(X_triain,y_train)
```
